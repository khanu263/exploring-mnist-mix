```python
import numpy as np
import random
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
dataset = np.load('MNIST-MIX-all/MNIST_MIX_train_test.npz')
X_train = dataset['X_train']
y_train = dataset['y_train']
```

```python
languages = {0: "arabic",
             1: "bangla",
             2: "devanagari",
             3: "english",
             4: "farsi",
             5: "kannada",
             6: "swedish",
             7: "telegu",
             8: "tibetan",
             9: "urdu"}
```

```python
label = 0
idx = 0
labels = []
images = []
while label < 100:
    if y_train[idx] == label:
        labels.append(label)
        images.append(X_train[idx])
        label = label + 1
    idx = idx + 1
num_col = 10
num_row = 10
fig, axes = plt.subplots(num_col, num_row, figsize=(1.5*num_col,2*num_row))
for i in range(100):
    ax = axes[i//num_col, i%num_col]
    ax.imshow(images[i], cmap='gray')
    ax.set_yticklabels([])
    ax.set_xticklabels([])
    ax.set_title("%s in %s" %(labels[i]%10,languages[int(labels[i]/10)]))
plt.tight_layout()
plt.show()
```
